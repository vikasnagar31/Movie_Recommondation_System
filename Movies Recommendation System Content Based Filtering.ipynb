```python
# now find unique genres  and apply one hot encoding to get features that are genres 
genres_list=list(data.genres.unique())
genres_list[:5]
```

```python
unique_genres=set()
for i in genres_list:
    unique_genres.update(i.split('|'))
     
print(sorted(unique_genres))
```

```python
#convert unique geners to list because set is not mutable 
unique_genres=list(unique_genres)
type(unique_genres)
```

```python
   # or  
unique_genres=[]
 
for i in genres_list:
    li=i.split('|')
    
    for j in li:
        if j not in unique_genres:
            unique_genres.append(j)
print(unique_genres)

# we use it because set is not mutable
```

```python
data[data.genres=='(no genres listed)'].shape
```

```python
df=data[data.genres!='(no genres listed)'] 
df.reset_index(drop=True,inplace=True)
```

```python
df  # now we got preocessed data and it is ready to precess for content based filtering 
```

```python
unique_genres.remove('(no genres listed)')
print(unique_genres)
```

```python
#this is like a base to fill the genre values for final_df so let go ahed  
genres_df=pd.get_dummies(unique_genres).astype('int64')
genres_df
```

```python
genres_df=genres_df[genres_df.index==0]
genres_df        

# this step is mostly used of keeping a single row
```

```python
genres_df['Adventure']=0
genres_df
```

```python
df2=pd.concat([df,genres_df],axis=1)
df2.head()
```

```python
df2.fillna(0,inplace=True)
df2.head()       

# we do that stuff to get all values 0 so we will add new values to it 
```

####  keep attention on next line 

```python
# before apply it we have to understand the next cell 

def create_encoding(tmp_df):
    genre_l = tmp_df['genres']   # generally it returns a series but we will apply the function row wise it 
    #print(genre_l)
    genre_l = genre_l.split('|')
    #print(genre_l)
    for g in genre_l:
        tmp_df[g]=1
        #display(tmp_df)
    return tmp_df 
 
```

```python
final_df = df2.apply(lambda x: create_encoding(x),axis=1)     # axis=1 means apply function to each row of genre column.
#Here, x is each row (a Pandas Series).
#create_encoding(x) is called on one row at a time, meaning x['genres'] is a string.
final_df.head()
```

```python
final_df.columns
```

```python
movie_df=final_df[['movieId','year', 'Action', 'Adventure',
       'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'IMAX', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
movie_df.head(3)
```

```python
movie_df['movieId'].values
```

```python
movie_df.index=movie_df['movieId'].values  # cosine similarity create n*n dataframe based on index of dataframe that we pass in cosine_similarity
```

```python
movie_df.drop(columns=['movieId'],inplace=True)
```

```python
movie_df      # it includes all the features
```

```python
# getting the similarities betweeen movies, here data is at row level,  as high the the values as similar the movie
movie_similarity=cosine_similarity(movie_df,movie_df)
movie_similarity=pd.DataFrame(movie_similarity)
movie_similarity
```

```python
movieId_mapping=dict(zip(movie_similarity.index.values,movie_df.index.values))
movieId_mapping
```

```python
# to make the index relavant that show the true relation as data is at row level where each row represent one single movie and it's similarity with others 
movie_similarity.index=movie_df.index.values
movie_similarity.columns=movie_df.index.values
movie_similarity
```

```python
# for reference for understanding purpose

#movies_sim_score = movie_similarity[movie_similarity.index==2]
#np.argsort(movies_sim_score.values)
 
```

```python
for i in [1, 2, 3]:  # Here we are passing movie IDs
    #print('Recommended movies for movieId', i)
    movies_id_score = movie_similarity[movie_similarity.index == i]
    top_3_similar_movies = list(np.argsort(movies_id_score.values)[0])[-4:-1][::-1]
    #print('Movie Index:',top_3_similar_movies)

    similar_movies = [movieId_mapping[j] for j in top_3_similar_movies]
    
    print(f'Similar movies for movie {i} are:\n', similar_movies)
```

```python
for i in final_df.movieId.values[0:100]:   
    print('MovieId', i)
    display(final_df[final_df.movieId == i][['movieId', 'title', 'genres']])

    movie_id_score = movie_similarity[movie_similarity.index == i]
    top_3_similar_movies = list(np.argsort(movie_id_score.values)[0])[-4:-1][::-1]   

    similar_movies = [movieId_mapping[j] for j in top_3_similar_movies]

    print(f'Similar movies for MovieId {i} are:')
    display(final_df[final_df.movieId.isin(similar_movies)][['movieId', 'title', 'genres','year']])
    print('------------------------------------------------------------------------------')

    
```
